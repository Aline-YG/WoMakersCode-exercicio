#Desafio 1 - Importando dataset pelo pydataset

Definição do problema: Dado o código do dataset retorne as seguintes informações:

    Importe o dataset utilizando a seguinte função do pydataset: data(“Código”)
    Imprimir na tela o dataset;
    Informe o tipo de dados retornado pela função data;
    Informe o número de exemplos (linhas) e características (colunas) do dataset.
    Crie uma função que ao receber um DataFrame retorna o número de linhas e colunas.

Código do dataset: plantTraits

Dicas: plantTraits é um dataset que contém a descrição, por meio de atributos biológicos, de diferentes espécies de plantas; função built-in (nativa da linguagem Python): type() atributo do DataFrame: shape

```python
# Importanto dataset
from pydataset import data
plantes = data('plantTraits')
plantes.head()
```

```python
# Iprimindo o dataset ma tela
print(plantes)
```

```python
# Mostrando os tipos de dados
type(plantes)
```

```python
print(type(plantes))
```

```python
# Informando quantidade de linhas e colunas
def numero_linhas_colunas(plantes):
  linhas = plantes.shape[0]
  colunas = plantes.shape[1]
  return linhas , colunas
linhas, colunas = numero_linhas_colunas(plantes)

print(f'O número de linhas é {linhas} e o número de colunas é {colunas}.')
```
